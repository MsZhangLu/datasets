```python
dummies = pd.get_dummies(df['productgroup_nm'], drop_first=True)
```

```python
df.head()
```

```python
df['units-1'] = df['units'].shift(1)
df['year'] = df['start_of_month'].dt.year
df['month'] = df['start_of_month'].dt.month
```

```python
train = df[df['start_of_month']<'2018-08-01'].copy()
test = df[df['start_of_month']=='2018-08-01'].copy()
```
